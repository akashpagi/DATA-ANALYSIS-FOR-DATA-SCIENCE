```python
import requests
import pandas as pd

# Create an empty list to store individual DataFrames
dfs = []

for i in range(1, 95):
    url = "https://api.themoviedb.org/3/tv/top_rated?api_key=2c2e435b3d23f2d65b06ce381da06caa&language=en-US&page={}".format(i)
    
    response = requests.get(url)
    
    # Check if 'results' key exists in the JSON response
    if 'results' in response.json():
        temp_df = pd.DataFrame(response.json()['results'])[['id','name','first_air_date','overview','popularity','vote_average','vote_count']]
        dfs.append(temp_df)
    else:
        print(f"No results found for page {i}")

# Concatenate all DataFrames in the list into one
df = pd.concat(dfs, ignore_index=True)

# Now df contains the aggregated data from all pages
```
