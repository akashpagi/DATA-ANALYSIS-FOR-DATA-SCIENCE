**What is a text file ?**

- A text file is a file that contains plain text without any special formatting or binary data.
- Text files typically consist of human-readable characters encoded in a standard character encoding, such as ASCII or UTF-8.
- They are used to store and exchange simple textual information and are commonly used for various purposes in computing.
- A text file identifies with the .txt file extension
