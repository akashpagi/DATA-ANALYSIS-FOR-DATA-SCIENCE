## S27.1 - WORKING WITH CSV FILES

```python
# importing pandas
import pandas as pd
```

#### **read_csv() function :**
- The read_csv() function in the pandas library is a powerful tool for reading tabular data from CSV files and creating a DataFrame, which is a two-dimensional, labeled data structure.
- **Syntax: pd.read_csv(filepath, sep, names, header='infer', index_col, usecols, dtype, converters, skiprows, nrows, encoding, na_values, parse_dates, chunksize)**


**What is a CSV FILE?**
- A CSV is a **comma-separated values** file, which allows data to be saved in a tabular format.
- CSVs look like a garden-variety spreadsheet but with a **.csv extension**.
- CSV files are mostly used for importing and exporting important information, such as customer or order data, to and from your database.
#### **Ways to importing csv files**

**1. opening the local csv files :**

```python
df = pd.read_csv('aug_train.csv')
df.head()
```

**2. opening a csv file from URL :**

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)
pd.read_csv(data)
```

**3. sep parameter:**
- It's specifies the separator or delimiter used in a CSV file to distinguish between fields.
- By default, it assumes a comma as the separator, but you can use sep to specify a different character or string, such as a semicolon (;) or tab ('\t').
- This is crucial for correctly parsing the data into a DataFrame, especially when the CSV file uses a different separator.

**4. names parameter:**
-  It's specify the column names for the resulting DataFrame.
-  It allows you to provide a list of names that will be used as column headers.

```python
# tsc : tab seperated values
pd.read_csv('movie_titles_metadata.tsv', 
            sep='\t',
            names=['sr.no','names','release_year','rating','votes','genres'])
```

**5. index_col parameter:**
-  This parameter used to specify which column should be used as the index (row labels) of the DataFrame. 

```python
# index_col parameters 
#  the 'enrollee_if' column from the CSV file is set as the index for the DataFrame. 
df = pd.read_csv('aug_train.csv',
                index_col='enrollee_id')
df.head()
```

**6. header parameter:**
- It is used to indicate which row of the CSV file should be considered as the header
-  By default, header is set to 0, if your data file doesn't have a header row, you can set header.

```python
df = pd.read_csv('test.csv', header=1)
df
```

**7. usecols parameter:**
- Take only selective columns form dataset

```python
df = pd.read_csv('aug_train.csv',
                usecols=['enrollee_id','gender','education_level'])
df.head()
```

**8. skipsrows parameter :**
- skipping the particular rows
- you can create own function to skipping the rows

```python
pd.read_csv('aug_train.csv',).head()
```

```python
pd.read_csv('aug_train.csv',skiprows=[3]).head()
```

**9. nrows parameter :**
- shows only specifed numbers of rows

```python
# nrows parameters
pd.read_csv('aug_train.csv', nrows=8)
```

**10. encoding parameter :**
- It is used to specify the character encoding of the CSV file you're reading. 
- It tells pandas how the characters in the file are represented.
- The default is usually 'utf-8', but you might need to specify a different encoding (like 'latin-1') if your file contains special characters or if the default encoding doesn't work.
- Choose an encoding that matches the character set of your data.

```python
pd.read_csv('zomato.csv', encoding='latin-1')
```

**11. on_bad_lines parameters (Skip bad line) :**
- If you want to skip lines in a CSV file that cause parsing errors, you can use the on_bad_lines parameter in the read_csv() function of pandas.
- Specifies what to do upon encountering a bad line (a line with too many fields). Allowed values are :
    -  ‘error’, raise an Exception when a bad line is encountered
    -   ‘warn’, raise a warning when a bad line is encountered and skip that lin
    -    ‘skip’, skip bad lines without raising or warning when they are encountered.

```python
pd.read_csv('BX-Books.csv', 
            sep=';', 
            encoding='latin-1',
            on_bad_lines='skip',)
```

**12. dtypes parameter :**
- To change the dtype of particular columns
- dtype or dict of {Hashable : dtype}

```python
pd.read_csv('aug_train.csv',dtype={'target': int}).info()
```

**13. parse_dates parameter (handling dates) :**
- Convert specified columns into datetime64 dtype
- If you are trying to parse multiple date columns, you can pass a list of column names to the parse_dates parameter.

```python
# Handling dates
pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

**14. converters parameter:**
- It specify a dictionary of functions for converting specific columns during the reading process.
- Each key in the dictionary represents a column name or column index, and the corresponding value is a function that will be applied to each element in that column.

```python
# creating function to replace the full form with short name
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
rename('Royal Challengers Bangalore')
```

```python
# converters with specific column
pd.read_csv('IPL Matches 2008-2020.csv',
            converters={'team1': rename}).head(3)
```

```python
# converters with multiple column list
columns_list = ['team1', 'team2', 'toss_winner', 'winner']
pd.read_csv('IPL Matches 2008-2020.csv',
            converters={ team:rename for team in columns_list}).head()
```

**15. na_values parameter (missing values):**
- Specified column or list of column values that should be treated as missing values (NaN) during the parsing of the CSV file.
- This parameter allows you to customize how missing or undefined values are identified.

```python
# na_values parameters
pd.read_csv('aug_train.csv',na_values=['Male'])
```

**16. chunksize parameter :**
-  loading a huge dataset in chunks
-  when you want to read a large CSV file in smaller, manageable chunks instead of loading the entire file into memory at once.
-  This is particularly useful when dealing with very large datasets that may not fit into memory.

```python
df = pd.read_csv('aug_train.csv', chunksize=4790)
```

```python
# here every chunks is small datatest
for chunks in df:
    print(chunks.shape)
```
